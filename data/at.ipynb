```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
table = pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv")
```

——— FORMAS DE ANALISAR E ILUSTRAR DADOS COM MATPLOTLIB ———

Quantidade de mulheres em cada categoria de peso:

```python
df_mulheres = table[table['Gender'] == 'Female']
contagem_peso = df_mulheres['NObeyesdad'].value_counts().sort_index()
plt.plot(contagem_peso.index, contagem_peso.values, marker='o', linestyle='-')
plt.title('Quantidade de Mulheres em Cada Categoria de Peso')
plt.xlabel('Categoria de Peso')
plt.ylabel('Quantidade')
plt.grid(True)
plt.xticks(rotation=45)  #rotaciona legendas no eixo x
plt.tight_layout() #ajustando sobreposição no layout
plt.show()
```

Quantidade de homens em cada categoria de peso:

```python
df_homens = table[table['Gender'] == 'Male']
contagem_peso_homens = df_homens['NObeyesdad'].value_counts().sort_index()
plt.bar(contagem_peso_homens.index, contagem_peso_homens.values, color='skyblue')
plt.title('Quantidade de Homens em Cada Categoria de Peso')
plt.xlabel('Categoria de Peso')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Comparação da quantidade de homens e mulheres para cada categoria de peso:

```python
contagem_peso_mulheres = table[table['Gender'] == 'Female']['NObeyesdad'].value_counts().sort_index()
contagem_peso_homens = table[table['Gender'] == 'Male']['NObeyesdad'].value_counts().sort_index()

contagem_peso = pd.DataFrame({
    'Mulheres': contagem_peso_mulheres,
    'Homens': contagem_peso_homens
}).fillna(0)  #substituindo valores NaN por 0

contagem_peso.plot(kind='bar', stacked=True, color=['pink', 'blue'])
plt.title('Distribuição de Categorias de Peso por Gênero')
plt.xlabel('Categoria de Peso')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.legend(title='Gênero')
plt.tight_layout()
plt.show()
```

Relação entre idade e peso:

```python
df_mulheres = table[table['Gender'] == 'Female']
df_homens = table[table['Gender'] == 'Male']

plt.figure(figsize=(10, 6))
plt.scatter(df_mulheres['Age'], df_mulheres['Weight'], color='pink', label='Mulheres', alpha=0.6)
plt.scatter(df_homens['Age'], df_homens['Weight'], color='blue', label='Homens', alpha=0.6)

plt.title('Relação entre Idade e Peso por Gênero')
plt.xlabel('Idade')
plt.ylabel('Peso')
plt.legend()
plt.grid(True)
plt.show()
```

Distribuição de peso por gênero:

```python
df_mulheres = table[table['Gender'] == 'Female']
df_homens = table[table['Gender'] == 'Male']

plt.figure(figsize=(10, 6))
plt.hist(df_mulheres['Weight'], bins=30, alpha=0.5, label='Mulheres', color='pink', edgecolor='black')
plt.hist(df_homens['Weight'], bins=30, alpha=0.5, label='Homens', color='blue', edgecolor='black')

plt.title('Distribuição de Peso por Gênero')
plt.xlabel('Peso')
plt.ylabel('Frequência')
plt.legend()
plt.grid(True)
plt.show()
```

Distribuição do peso geral:

```python
contagem_peso = table['NObeyesdad'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(contagem_peso, labels=contagem_peso.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title('Distribuição de Categorias de Peso')
plt.axis('equal')
plt.show()
```

Relação entre o peso e o meio de transporte:

```python
agrupado = table.groupby(['MTRANS', 'NObeyesdad']).size().unstack(fill_value=0)

agrupado.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribuição de Peso por Meio de Transporte')
plt.xlabel('Meio de Transporte')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.legend(title='Categoria de Peso')
plt.tight_layout()
plt.show()
```
